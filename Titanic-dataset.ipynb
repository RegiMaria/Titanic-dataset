```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Let's read our data in using pandas
```

```python
Titanic = pd.read_csv('titanic.csv')
Titanic
```

```python
#Specify the columns with data relevant to the analysis:
```

```python
df_Titanic= Titanic.loc[:, ['Name', 'Age','Sex','Pclass','Survived']]

df_Titanic
```

```python
#filtered the 'Sex' column for 'female' and the 'Survived' column equal to 1
```

```python
df_female_survivors = df_Titanic[(df_Titanic['Survived'] == 1) & (df_Titanic['Sex'] == 'female')]
df_female_survivors
```

```python
#Survivors by class:
```

```python
survivors_by_class= df_female_survivors.groupby('Pclass')['Survived'].sum()
survivors_by_class 
```

```python
#Let's take a Look at our data graphically:
```

```python
plt.title('Female Survivors by class')
survivors_by_class = df_female_survivors[df_female_survivors['Survived'] == 1]['Pclass'].value_counts().sort_values()
plt.bar(survivors_by_class.index, survivors_by_class.values, color='blue')
plt.xticks([1,2,3])
plt.xlabel('Class group')
plt.ylabel('Numbers of Survivors')
plt.savefig('df_female_survivors.png')
```
